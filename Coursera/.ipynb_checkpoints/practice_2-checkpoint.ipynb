```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
import requests

def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
```

```python
file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
download(file_path, "usedcars.csv")
file_name = "usedcars.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(file_name, names=headers)
df.head()
```

```python
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
missing_data = df.isnull()

for column in missing_data.columns:
    if missing_data[column].any(): 
        print(missing_data[column].value_counts())
        print("")
```

```python
replace_mean_col = {'normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm'}
replace_mean_col
```

```python
for i, column in enumerate(replace_mean_col):
    df[column] = df[column].replace(np.nan, df[column].astype(float).mean(axis=0))
    print(column)
```

```python
mod_num_of_doors = df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, mod_num_of_doors)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
try:
    df['highway-mpg'] = 235 / df['highway-mpg']
    df.rename(columns={'highway-mpg': 'highway-L/100km'}, inplace=True)
    print("Konversi dan rename kolom berhasil dilakukan.")
except KeyError:
    print("Operasi ini sepertinya sudah pernah dijalankan. Kolom sudah di-rename.")
```

```python
min_max_col = {'length', 'width', 'height'}
for i, column in enumerate(min_max_col):
    df[column] = df[column] / df[column].max()
```

```python
df.head()
```

```python
df['horsepower'] = df['horsepower'].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(df["horsepower"])
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ["low", "medium", "high"]
group_names
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
colors = ['lightcoral', 'skyblue', 'lightgreen']
pyplot.bar(group_names, df["horsepower-binned"].value_counts(), color=colors)

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
dum_var_1 = pd.get_dummies(df["fuel-type"])
dum_var_1
```

```python
dum_var_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'})
df.head()
```

```python
try:
    df.drop("fuel-type", axis=1, inplace=True)
    df = pd.concat([df, dum_var_1], axis=1)
except KeyError:
    print("proses concat selesai")
df.head()
```

```python
dum_var_2 = pd.get_dummies(df["aspiration"])
dum_var_2
```

```python
dum_var_2.rename(columns={"std":"aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)
dum_var_2
```

```python
try:
    df.drop("aspiration", axis=1, inplace=True)
    df = pd.concat([df, dum_var_2], axis=1)
except KeyError:
    print("concat-telah-dilakukan")
df.head()
```

```python
df.to_csv("clean_df.csv")
```
